```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/Student_Performance.csv')
```

```python
df.shape
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df.shape
```

```python
df = df.drop_duplicates(keep='first')
```

```python
df.shape
```

Clean Text Columns

```python
df['gender'] = df['gender'].str.strip().str.lower()
df['gender']
```

```python
df['school_type'] = df['school_type'].str.strip().str.lower()
df['school_type']
```

```python
df['parent_education'].value_counts()
```

```python
df['parent_education'] = df['parent_education'].str.strip().str.lower()
df['parent_education']
```

```python
df['travel_time'].value_counts()
```

```python
mapping = {
    '<15 min': 7.5,      # midpoint of 0-15
    '15-30 min': 22.5,   # midpoint of 15-30
    '30-60 min': 45,     # midpoint of 30-60
    '>60 min': 75        # assume average over 60
}

df['travel_time_min'] = df['travel_time'].map(mapping)
df['travel_time_min'].mean()
```

```python
df['final_grade'].value_counts()
```

```python
df['final_grade'].unique()
```

```python
df['final_grade'] = pd.Categorical(
    df['final_grade'],
    categories=['f','e','d','c','b','a'],  # from lowest to highest
    ordered=True
)
```

```python
df.to_csv('../data/processed/cleaned_data.csv', index=False)
```
