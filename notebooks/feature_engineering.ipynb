```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
import joblib
```

```python
df = pd.read_csv('../data/processed/cleaned_data.csv')
```

Categorical Columns:

gender
- Gender is categorical
- One hot encoding
- No ordinal relationship exists between categories

```python
df = pd.get_dummies(df, columns=['gender'], drop_first=True, dtype=int)
```

```python
df.head()
```

school_type
- Only two categories, so it’s binary
- Map it to 0/1 directly

```python
df['school_type_encoded'] = df['school_type'].map({'public':0,'private':1})
```

parent_education
- Categorical and ordinal column
- Ordinal encoding

```python
education_order = [['no formal', 'high school', 'diploma', 'graduate', 'post graduate', 'phd']]
encoder = OrdinalEncoder(categories=education_order)

df['parent_education_encoded'] = encoder.fit_transform(df[['parent_education']])
```

study_hours
- Numerical Column

attendance_percentage
- Numerical column

internet_access
- Categoical & binary
- Map it to 0/1 directly

```python
df['internet_access_encoded'] = df['internet_access'].map({'yes':1,'no':0})
```

extra_activities
- Categorical & binary column
- 0/1 encoding

```python
df['extra_activities_encoding'] = df['extra_activities'].map({'yes':1,'no':0})
```

study_method
- Categorical & no order
- One hot encoding

```python
df = pd.get_dummies(df,columns=['study_method'],drop_first=True,dtype=int)
```

```python
df[df['student_id'] == 6]
```

final_grade
- categorical & ordinal
- label encoding

```python
encoder = LabelEncoder()
df['final_grade_encoded'] = encoder.fit_transform(df['final_grade'])
```

```python
df.head()
```

Drop Unnecessary Columns from the Dataset

```python
df.drop(columns=['student_id','school_type','parent_education','internet_access','travel_time','extra_activities','final_grade'],inplace=True)
```

Rename columns

```python
df.rename(columns={'study_method_group study': 'study_method_group_study'}, inplace=True)
df.rename(columns={'study_method_online videos': 'study_method_online_videos'}, inplace=True)
```
